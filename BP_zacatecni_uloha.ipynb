- generovat N realizaci nahodne veliciny s rozdelenim N(0,1)

```python
mu, sigma = 0, 1 # mean and standard deviation
s = np.random.normal(mu, sigma, 1000)
```

- graf histogramu z generovanych dat a presne hustoty pravdepodobnosti (PDF)

```python
import matplotlib.pyplot as plt
#zobrazeni hist.
count, bins, ignored = plt.hist(s, 30, density=True)
#zobrazeni pdf
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *  
               np.exp( - (bins - mu)**2 / (2 * sigma**2) ), 
         linewidth=2, color='r')
plt.show()
```

- graf empiricke distribucni funkce (ECDF) a presne distribucni funkce (CDF)

```python
x = np.sort(s)
y = np.arange(1, len(x) + 1) / len(x)
z = plt.plot(x, y, marker='.', linestyle='none')
z = plt.xlabel('s')
z = plt.ylabel('ECDF')
#CDF = plt.plot
plt.margins(0.02) # keeps data off plot edges
plt.show()
```

  C je korelacni matice s hodnotou C = R^T D R, kde R je rotacni matice
  R = 1/sqrt(2) * [[1, 1], [-1, 1]] 
  a D je diagonalni skalovaci matice D = [[ 10, 0], [0, 1]]

```python
mean = [1, 2]
import math # calculating corelation matrix
r = [1, 1],[-1, 1]
r_array = np.array(x)
R = 1/math.sqrt(2) * r_array 
d = [[ 10, 0], [0, 1]]
D = np.array(d)
Rt = np.transpose(R)
C = Rt * D * R # diagonal covariance
```
